```python
import re

def clean_text(text):
    # Remove multiple newlines and replace with a single one
    text = re.sub(r'\n\s*\n', '\n\n', text)
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    # Strip leading/trailing whitespace from each line and the whole text
    text = '\n'.join([line.strip() for line in text.split('\n')])
    text = text.strip()
    return text

processed_text = clean_text(extracted_text)

print("--- First 1000 characters of processed text ---")
print(processed_text[:1000])
print("-----------------------------------------------")
print(f"Original text length: {len(extracted_text)}")
print(f"Processed text length: {len(processed_text)}")
```

```python
def generate_question_paper(notes):
    # ✅ Define prompt FIRST
    prompt = f"""
    Create a question paper from the notes.

    Section A: 5 questions (2 marks, easy)
    Section B: 3 questions (5 marks, medium)
    Section C: 2 questions (10 marks, hard)

    Notes:
    {notes}
    """

    # ✅ Call Groq API
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    return response.choices[0].message.content
```

```python
notes = "Machine learning is a subset of AI..."

paper = generate_question_paper(notes)

print(paper)
```
